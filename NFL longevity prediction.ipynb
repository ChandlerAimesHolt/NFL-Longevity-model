```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

# Set up Selenium WebDriver
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # Runs Chrome in the background
driver = webdriver.Chrome(options=options)

# Positions & Years to Scrape
positions = ['DE', 'DT']
years = range(1987, 2019)  # From 1987 to 2018

# List to store all DataFrames
all_data = []

# Loop through each year & position
for year in years:
    for pos in positions:
        url = f'https://nflcombineresults.com/nflcombinedata.php?year={year}&pos={pos}&college=&proday=off'
        print(f"Fetching data for {pos} in {year}...")

        # Load the page
        driver.get(url)

        # Wait for the table to load
        try:
            WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.TAG_NAME, 'table'))
            )
        except:
            print(f"Skipping {pos} in {year} - Table did not load.")
            continue  # Skip if table doesn't load

        # Extract HTML after full load
        html = driver.page_source

        # Use Pandas to parse the table
        dfs = pd.read_html(html)
        
        if dfs:
            df = dfs[0]  # Assuming the first table contains the data
            df['Year'] = year  # Add Year Column
            df['Position'] = pos  # Add Position Column
            all_data.append(df)
        else:
            print(f"No table found for {pos} in {year}.")

# Close WebDriver
driver.quit()

# Combine all DataFrames
NFL_Combine_Results = pd.concat(all_data, ignore_index=True)
```

```python
import re
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

# Load your DataFrame
df = NFL_Combine_Results  # Replace with actual DataFrame

# Path to ChromeDriver (update as needed)
chrome_driver_path = r"C:\Users\chand\Documents\SQL data\Python Data\chromedriver-win64\chromedriver.exe"

# Set up Selenium WebDriver in headless mode
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run in background
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")

service = Service(chrome_driver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)

# Create a new column for RAS Scores
df["RAS"] = None

# Base URL for the RAS calculator
base_url = "https://ras.football/ras-calculator/"

# Loop through every row in the dataset
for index, row in df.iterrows():
    try:
        print(f"Processing {row['Name']}...")

        # Generate the full URL with parameters
        full_url = f"{base_url}?pos={row['POS']}&ht={int(row['FIIE'])}&wt={int(row['Weight (lbs)'])}&ft={row['40 Yard']}&bn={row['Bench Press']}&vt={row['Vert Leap (in)']}&bd={row['Broad Jump (in)']}&sh={row['Shuttle']}&cn="

        print(f"Generated URL: {full_url}")

        # Open the URL
        driver.get(full_url)

        # Wait for the page to load completely
        time.sleep(3)

        # Get the full page source
        page_source = driver.page_source

        # Use regex to find the RAS score (digits before "u<")
        match = re.search(r"(\d+\.\d+)\s*u<", page_source)

        if match:
            ras_score = match.group(1)  # Extract the matched number
            df.at[index, "RAS"] = ras_score
            print(f"✅ {row['Name']} - RAS Score: {ras_score}")
        else:
            print(f"⚠️ RAS Score not found for {row['Name']}")

        # Sleep to avoid excessive requests
        time.sleep(2)

    except Exception as e:
        print(f"❌ Error processing {row['Name']}: {e}")

# Close the browser
driver.quit()

# Save the updated DataFrame
df.to_csv("RAS_results.csv", index=False)

```

```python
# Function to format the player name into a valid URL path
def format_player_url(name):
    formatted_name = name.lower().replace(" ", "-").replace("'", "")  # Convert to lowercase, replace spaces, remove apostrophes
    return f"https://www.nfl.com/players/{formatted_name}/"

# Apply the function to generate URLs for each player in df_cleaned
df_cleaned["Profile_URL"] = df_cleaned["Name"].apply(format_player_url)
```

```python
import requests
from bs4 import BeautifulSoup
import time
from tqdm import tqdm

# Function to extract experience and handle non-numeric values
def get_experience(url):
    try:
        response = requests.get(url)
        response.raise_for_status()  # Ensure the request was successful
        soup = BeautifulSoup(response.text, 'html.parser')
        experience_key = soup.find('div', class_='nfl-c-player-info__key', string='Experience')
        
        if experience_key:
            experience_value = experience_key.find_next_sibling('div', class_='nfl-c-player-info__value')
            if experience_value:
                exp_text = experience_value.text.strip()
                
                # If the value is numeric, convert to int, otherwise return 0
                return int(exp_text) if exp_text.isdigit() else 0  
        
        return None
    except requests.exceptions.RequestException as e:
        print(f"Request failed: {e}")
        return None

# Ensure df_cleaned exists and contains 'Profile_URL'
if "df_cleaned" in locals() and "Profile_URL" in df_cleaned.columns:
    df_cleaned["Experience"] = None  # Initialize the column

    # Iterate over each row to fetch experience data
    for index, row in tqdm(df_cleaned.iterrows(), total=df_cleaned.shape[0]):
        url = row["Profile_URL"]
        df_cleaned.at[index, "Experience"] = get_experience(url)
        time.sleep(1)  # Delay to avoid blocking


else:
    print("Error: df_cleaned does not exist or is missing 'Profile_URL'.")
```
